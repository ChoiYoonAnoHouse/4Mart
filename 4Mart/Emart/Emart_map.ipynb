```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import pandas as pd
import matplotlib.font_manager as fm
import folium
from matplotlib import rc
import base64
```

```python
Emart_df = pd.read_csv('./Emart_info.csv',encoding='utf-8')
Emart_df.head()
```

```python
map = folium.Map(location=[35.2500198,129.0907241],zoom_start=16,tiles='OpenStreetMap')
folium.Marker(location=[35.2500198,129.0907241],icon=folium.Icon(color='gray',icon='shopping-cart'),tooltip='Emart').add_to(map)
map
```

```python
map = folium.Map(location=[35.2500198,129.0907241],zoom_start=10,tiles='OpenStreetMap')

for i in Emart_df.index:
    lt_lat = Emart_df.loc[i,'Latitude']
    lt_long = Emart_df.loc[i,'Longitude']

    title = f"{Emart_df.loc[i, 'store']}"
    
    pic = base64.b64encode(open('Emart_Logo.png','rb').read()).decode()
    image_tag = f'<img src="data:image/png;base64,{pic}"><br><p>{title}</p>'
    iframe = folium.IFrame(image_tag, width=350, height=100)
    popup = folium.Popup(iframe, max_width=650)

    color = 'gray'
    if Emart_df.loc[i,'store']=='Emart':
        color ='gray'

    folium.Marker(location=[lt_lat,lt_long],radius=5,color=color,icon=folium.Icon(icon='shopping-cart',color='gray'),popup=popup).add_to(map)
map
```
