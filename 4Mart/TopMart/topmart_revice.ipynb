```python
topmart = pd.read_csv('./topmart_shop_info_loc.csv', index_col=0, encoding='utf-8')
topmart.head()
```

```python
addr = []
for address in topmart.address:
    addr.append(str(address).split())
addr
```

```python
addr2 = []

for i in range(len(addr)):
    if addr[i][0] == '부산': addr[i][0] = '부산광역시'
    elif addr[i][0] == '경남': addr[i][0] = '경상남도'
    elif addr[i][0] == '경북': addr[i][0] = '경상북도'
    elif addr[i][0] == '울산': addr[i][0] = '울산광역시'
    elif addr[i][0] == '대구': addr[i][0] = '대구광역시'
    
    addr2.append(' '.join(addr[i]))
    
print(addr2)
print(len(addr2))
```

```python
addr2 = pd.DataFrame(addr2, columns=['address2'])
```

```python
topmart = pd.concat([topmart,addr2], axis=1)
topmart.head()
```

```python
topmart = topmart.drop('address', axis=1)
```

```python
topmart = topmart.rename(columns={'address2':'address'})
```

```python
topmart = topmart[['store', 'address', 'Latitude', 'Longitude', 'tel']]
topmart
```

```python
topmart.to_csv('./topmart_shop_info_loc.csv', encoding='utf-8', index=False)
topmart.head()
```
